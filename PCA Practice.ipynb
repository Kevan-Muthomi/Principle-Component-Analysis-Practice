# Principle Component Analysis

- A proper statistical technique used in recognition classification
- Effective for revealing necessary data information
- Has limitations when handling large datasets
- Reduces Overfitting
- Independent variables become less interpretable

### Importing Libraries 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
digits = load_digits()
```

### Reading the data

```python
data = pd.DataFrame(digits['data'],columns = digits['feature_names'])
data.head()
```

```python
data.tail()
```

## PCA Preparation 

### Standardization

```python
#Standardize the data
scale = StandardScaler()

scaled_data = scale.fit_transform(data)

#Create a new Data Frame
data_new = pd.DataFrame(scaled_data, columns =digits['feature_names'])

data_new.head()
```

### PCA With 2 Components

```python
pca = PCA(n_components=2)

#Tranform the data
data_pca = pca.fit_transform(data_new)

#Checking if changes have taken effect
data_pca.shape

#Creating a new dataframe
component_data = pd.DataFrame(data=data_pca, columns=['First Component','Second Component'])
component_data.head()
```
